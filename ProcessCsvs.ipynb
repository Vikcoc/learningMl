```python
from datetime import datetime
df['Timestamp'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d %H:%M:%S.%f'))
print(df['Timestamp'])
```

```python
df["Timestamp"] = pd.to_datetime(df['Timestamp'], errors='coerce')
```

```python
df["RoundedTimestamp"] = df["Timestamp"].dt.floor("min")
```
